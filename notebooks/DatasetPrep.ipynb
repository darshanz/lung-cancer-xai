```python
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class LungOneDataset(Dataset):
    def __init__(self, fold=0, train=True):
        DATA_DIR  = "/Users/sudarshan/darshanz/datasets/lung1"
        label_df = pd.read_csv(f'{DATA_DIR}/label.csv')
        self.data_  = label_df[label_df['fold'] != fold__] if train else label_df[label_df['fold'] == fold__]
        self.event_status =  data_['event_status']
        self.survival_time = data_['survival_time'] 
        self.len = data_['event_status'].shape[0] 
        
    def __getitem__(self, index):
        patient_id = self.data_.iloc[index]['patient_id']
        ct = torch.load(f'{DATA_DIR}/CT_ONLY/vols/{patient_id}.pt')
        return ct, data_.iloc[index]['survival_time'], data_.iloc[index]['event_status']
        
    def __len__(self):
        return self.len
```

```python
train_data = LungOneDataset(fold=0)
test_data = LungOneDataset(fold= 0, train=False) 
trainloader = DataLoader(train_data, batch_size=4)
testloader = DataLoader(test_data, batch_size=4)
```
